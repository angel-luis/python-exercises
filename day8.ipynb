Write a program to compute the frequency of the words from the input. The output should output after sorting the key alphanumerically.

Suppose the following input is supplied to the program:
```
New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3.
```

Then, the output should be:
```
2:2
3.:1
3?:1
New:1
Python:5
Read:1
and:1
between:1
choosing:1
or:2
to:1
```

````python
user_input = input("Write a phrase... ").split(
    " ")

frequency_list = {}
for unique_word in sorted(set(user_input)):
    frequency_list[unique_word] = user_input.count(unique_word)

for word, freq in frequency_list.items():
    print(f"{word}:{freq}")
````

Write a method which can calculate square value of number

````python
user_input = int(input("Write a num"))

def square(num):
    return num**2

square(user_input)
````

Please write a program to print some Python built-in functions documents, such as abs(), int(), raw_input()

And add document for your own function

````python
user_input = input("Write a function... ")

def show_help(fn):
    """
    Returns the documentation of the function provided
    """
    print(help(fn))

show_help(user_input)
print(show_help.__doc__)
````
